# Sécurité et conditions d’utilisation de l’application Streamlit Copilot Adoption Dashboard

Ce notebook présente les mesures de sécurité et les conditions d’utilisation de l’application Streamlit développée pour l’analyse de l’adoption de Copilot. Il est destiné à un usage interne au sein de VINCI Energies.


## 1. Introduction

L’application Streamlit permet aux utilisateurs d’importer des fichiers CSV exportés depuis M365 pour générer des tableaux de bord d’adoption. Ce document vise à informer les utilisateurs sur les pratiques de sécurité et de confidentialité à respecter.


## 2. Fonctionnement général de l’application

- L’utilisateur importe ses propres fichiers CSV via l’interface Streamlit.
- Les données sont traitées uniquement en mémoire pendant la session.
- Aucun stockage permanent des données n’est réalisé sur le serveur ou le poste.
- Les exports (CSV) sont générés à la demande et téléchargés localement.


## 3. Sécurité des données

- Les fichiers importés ne sont pas sauvegardés sur le disque.
- Les données sont effacées à chaque fermeture ou rafraîchissement de la session.
- L’application ne stocke ni mots de passe, ni tokens, ni secrets.
- Les fichiers générés sont téléchargés par l’utilisateur et ne restent pas sur le serveur.


## 4. Confidentialité

- L’application ne conserve aucune information personnelle ou confidentielle après la session.
- Aucun partage automatique des données avec des tiers.
- Les logs utilisateurs sont visibles uniquement par l’utilisateur qui importe les données.


## 5. Bonnes pratiques recommandées à l’utilisateur

- Ne pas partager l’accès à l’application avec des personnes non autorisées.
- Vérifier que les fichiers importés ne contiennent pas d’informations sensibles inutiles.
- Télécharger les exports CSV uniquement sur des postes sécurisés.
- Ne pas versionner de fichiers contenant des données personnelles ou confidentielles.


## 6. Limites et recommandations complémentaires

- Si l’application doit être partagée sur un réseau, ajouter une authentification (ex : Streamlit-Authenticator).
- Pour un usage en production, activer le chiffrement des échanges (HTTPS).
- Respecter la politique de sécurité informatique de VINCI Energies.


## 7. Conditions d’utilisation

- Usage strictement interne VINCI Energies.
- L’utilisateur est responsable des données qu’il importe.
- L’application ne doit pas être utilisée pour traiter des données à caractère personnel sans autorisation.
